```python

# Abrir el navegador Edge
driver = webdriver.Edge()

links=[]

try:
    # Página web
    url = 'https://www.argenprop.com/departamentos/alquiler/resistencia'

    # Abrir la página
    driver.get(url)

    # Obtener el elemento entre pagination__separator y pagination__page-next
    elemento = driver.find_element(By.XPATH, '//li[@class="pagination__separator"]/following-sibling::li[1][not(contains(@class, "pagination__page-next"))]')

    # Convertir el texto del elemento a un número entero
    numero_paginas = int(elemento.text)

    for i in range(1, numero_paginas + 1):
        print(f"Página {i}")
        
        # Obtener el div que contiene los departamentos
        try:
            departamentos_div = driver.find_element(By.XPATH, '//div[@class="listing__items"]')
        except NoSuchElementException:
            print("No se encontró el div que contiene los departamentos")
            continue
        
        # Obtener los departamentos usando un XPath relativo
        departamentos = departamentos_div.find_elements(By.XPATH, './div')
        print(f"Número de departamentos: {len(departamentos)}")
        
        for f in range(len(departamentos)):
            print(f"Departamento {f+1}")
            try:
                enlaces = departamentos[f].find_elements(By.XPATH, './/a')
                for enlace in enlaces:
                    href = enlace.get_attribute('href')
                    print(f"Href: {href}")
                    links.append(href)
            except NoSuchElementException:
                print(f"No se encontró el enlace en el departamento {f+1}")
        
        # Cambiar de página
        if i < numero_paginas:
            try:
                siguiente = driver.find_element(By.XPATH, '//li[@class="pagination__page-next pagination__page"]/a')
                siguiente_href = siguiente.get_attribute('href')
                driver.get(siguiente_href)
                time.sleep(3)  # Esperar un momento para que la página cargue
            except NoSuchElementException:
                print("No se encontró el botón de siguiente página")
                break
finally:
    # Cerrar el navegador
    driver.quit()
```

```python
# iterar sobre los enlaces guardados en la lista links
data_argenprop = []


driver = webdriver.Edge()
try:
    for enlace in links:

        driver.get(enlace)
        time.sleep(3)  # Esperar un momento para que la página cargue
        
        # Obtener la información del departamento
        try:
            address = driver.find_element(By.XPATH, '//h2[@class="titlebar__address"]').text
        except NoSuchElementException:
            print("No se encontró la direccion")
            address = None
        
        try:
            price = driver.find_element(By.XPATH, '//p[@class="titlebar__price"]').text
        except NoSuchElementException:
            print("No se encontró el precio")
            price = None
        
        
        #div de descripcion
        try:
            expensas = driver.find_element(By.XPATH, '//p[@class="titlebar__expenses hide-in-mobile"]').text
        except NoSuchElementException:
            print("No se encontró el precio de las expensas")
            expensas = None
        
        try:
            descripcion = driver.find_element(By.XPATH, '//div[@class="property-description"]/ul[@class="property-main-features"]')
        except NoSuchElementException:
            print("No se encontró la descripción")
            descripcion = None
            
            
        try:
            tipo = descripcion.find_element(By.XPATH, './li[1]').text
        except NoSuchElementException:
            print("No se encontró el tipo")
            tipo = None
            
        try:
            superficie = descripcion.find_element(By.XPATH, './li[@title="Sup. cubierta"]').text
        except NoSuchElementException:
            print("No se encontró la superficie")
            superficie = None
            
        try:
            cantidad_ambientes = descripcion.find_element(By.XPATH, './li[@title="Ambientes"]').text
        except NoSuchElementException:
            print("No se encontró la cantidad de ambientes")
            cantidad_ambientes = None
            
        try:
            cantidad_dormitorios = descripcion.find_element(By.XPATH, './li[@title="Dormitorios"]').text
        except NoSuchElementException:
            print("No se encontró la cantidad de dormitorios")
            cantidad_dormitorios = None
            
        try:
            cantidad_banos = descripcion.find_element(By.XPATH, './li[@title="Baños"]').text
        except NoSuchElementException:
            print("No se encontró la cantidad de baños")
            cantidad_banos = None
            
        try:
            # obtener div de mapa de google map
            mapa = driver.find_element(By.XPATH, '//div[@class="leaflet-container leaflet-fade-anim leaflet-grab leaflet-touch-drag"]')
        except NoSuchElementException:
            print("No se encontró el mapa")
            
        
        #obtener latitud y longitud de los atributos del div
        try:
            latitud = mapa.get_attribute('data-latitude')
        except NoSuchElementException:
            print("No se encontró la latitud")
            latitud = None
            
        try:
            longitud = mapa.get_attribute('data-longitude')
        except NoSuchElementException:
            print("No se encontró la longitud")
            longitud = None
            
        departamento_info = {
            "enlace": enlace,
            "address": address,
            "price": price,
            "expensas": expensas,
            "tipo": tipo,
            "superficie": superficie,
            "cantidad_ambientes": cantidad_ambientes,
            "cantidad_dormitorios": cantidad_dormitorios,
            "cantidad_banos": cantidad_banos,
            "latitud": latitud,
            "longitud": longitud
        }
        
        print(departamento_info)     
        
        data_argenprop.append(departamento_info)   



finally:
    driver.quit()
```

```python
#lista a data frame
df_argenpro = pd.DataFrame(data_argenprop)
df_argenpro.to_csv('departamentos_argenprop.csv', index=False)
```

```python
import re
def extraer_latitud_longitud(url):
    # Expresión regular para capturar la latitud y longitud
    patron = r"!3d(-?\d+\.\d+)!4d(-?\d+\.\d+)"
    coincidencia = re.search(patron, url)
    
    if coincidencia:
        latitud = float(coincidencia.group(1))
        longitud = float(coincidencia.group(2))
        return latitud, longitud
    else:
        return None
```

```python
# Abrir el navegador Edge
driver = webdriver.Edge()

links=[]

try:
    # Página web
    url = 'https://www.zonaprop.com.ar/casas-departamentos-ph-alquiler-resistencia.html'

    # Abrir la página
    driver.get(url)

    
    # Obtener el elemento entre pagination__separator y pagination__page-next
    try:
        wait = WebDriverWait(driver, 10)
        texto_con_cantidad = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div/div/div[3]/div/div[1]').text
    except NoSuchElementException:
        print("No se encontró el texto con la cantidad de departamentos")
    # extraer el texto con la cantidad de departamentos dejando solo los numeros
    numero_despartamentos = int(''.join(filter(str.isdigit, texto_con_cantidad)))

    numero_paginas = int(np.ceil(numero_despartamentos / 30))

    print(f"Numero de paginas: {numero_paginas}")



    for i in range(1, numero_paginas + 1):
        print(f"Página {i}")
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class="postings-container"]')))


        # Obtener el div que contiene los departamentos
        try:
            departamentos_div = driver.find_element(By.XPATH, '//div[@class="postings-container"]')
        except NoSuchElementException:
            print("No se encontró el div que contiene los departamentos")
            continue
        
        # Obtener los departamentos usando un XPath relativo
        departamentos = departamentos_div.find_elements(By.XPATH, './div')
        print(f"Número de departamentos: {len(departamentos)}")
        
        for f in range(len(departamentos)):
            print(f"Departamento {f+1}")
            WebDriverWait(driver, 10)

            try:
                enlaces = departamentos[f].find_elements(By.XPATH, './/a')
                for enlace in enlaces:
                    href = enlace.get_attribute('href')
                    print(f"Href: {href}")
                    links.append(href)
            except NoSuchElementException:
                print(f"No se encontró el enlace en el departamento {f+1}")

        # Cambiar de página
        if i < numero_paginas:
            try:
                WebDriverWait(driver, 10)

                siguiente = driver.find_element(By.XPATH, '//a[@class="PageArrow-sc-n5babu-2 kTvCSV"]')
                siguiente_href = siguiente.get_attribute('href')
                driver.get(siguiente_href)
                time.sleep(3)  # Esperar un momento para que la página cargue
            except NoSuchElementException:
                print("No se encontró el botón de siguiente página")
                break
                
        time.sleep(random.uniform(1, 4))

finally:
    # Cerrar el navegador
    # driver.quit()
    print("fin")
```

```python
import re

def obtener_latitud(url):
    # Expresión regular para capturar la latitud
    patron_latitud = r"!3d(-?\d+\.\d+)"
    coincidencia_latitud = re.search(patron_latitud, url)
    print(coincidencia_latitud)
    
    if coincidencia_latitud:
        latitud = float(coincidencia_latitud.group(1))
        print(latitud)
        return latitud
    else:
        return None

def obtener_longitud(url):
    # Expresión regular para capturar la longitud
    patron_longitud = r"!2d(-?\d+\.\d+)"
    coincidencia_longitud = re.search(patron_longitud, url)
    print(coincidencia_longitud)
    
    if coincidencia_longitud:
        longitud = float(coincidencia_longitud.group(1))
        print(longitud)
        return longitud
    else:
        return None
```
