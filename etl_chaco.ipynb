```python
#cambia mono por 1

import pandas as pd
import re

# Expresión regular para detectar "mono" o "monoambiente" en todas sus variantes
regex_exp = re.compile(r'\bmono(?:[-\s]?ambiente)?\b', re.IGNORECASE)

# Aplicar la expresión regular a la columna 'cantidad_ambientes' y cambiar los valores que cumplen la expresión regular por 1
df['cantidad_ambientes'] = df['cantidad_ambientes'].apply(lambda x: '1' if isinstance(x, str) and regex_exp.search(x) else x)
```

```python
# transformar los datos de estring a numericos
df['price'] = df['price'].str.replace(r'[^\d]', '', regex=True)
df['expensas'] = df['expensas'].str.replace(r'[^\d]', '' , regex=True)
df['cantidad_dormitorios'] = df['cantidad_dormitorios'].str.replace(r'[^\d]', '', regex=True)
df['cantidad_banos'] = df['cantidad_banos'].str.replace(r'[^\d]', '', regex=True)
df['cantidad_ambientes'] = df['cantidad_ambientes'].str.replace(r'[^\d]', '', regex=True)
df['superficie'] = df['superficie'].str.replace(r'[^\d]', '', regex=True)

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['expensas'] = pd.to_numeric(df['expensas'], errors='coerce')
```

```python
data_universidades = {
    'nombre': ['UNNE', 'UTN','UCES','UCuencaPlata', 'UCSalta','UGastónDachary', 'US21'],
    'latitud': [-27.4634275, -27.4511154 , -27.4495898, -27.4594811,-27.4559077, -27.455657, -27.4526272],
    'longitud': [-58.9836224, -58.9790106 , -58.9810485, -58.9855027, -58.9843114, -58.9850876 ,-58.9929929]
}
df_universidades = pd.DataFrame(data_universidades)

df_universidades_nombre = df_universidades['nombre']
df_universidades_nombre.to_csv(r'dataset\postETL\universidades_chaco.csv' , index=False)
```

```python
import pandas as pd
import numpy as np

# Función para calcular la distancia en cuadras entre dos puntos dados sus coordenadas de latitud y longitud
def calcular_distancia_cuadras(lat1, lon1, lat2, lon2):
    # Si lat1 o lon1 son NaN, asignarles los valores de lat2 y lon2
    
    
    if pd.isnull(lat1) or pd.isnull(lon1):
        return np.nan

    # Convertir las coordenadas a radianes
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Calcular la diferencia entre las coordenadas
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Calcular la distancia en metros usando la fórmula de Haversine
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    distancia_metros = 6371000 * c  # Radio de la Tierra en metros
    
    # Convertir la distancia a cuadras (1 cuadra = 100 metros)
    distancia_cuadras = distancia_metros // 100
    return int(distancia_cuadras)
    

# Calcular la distancia de cada departamento a cada universidad
# Calcular la distancia de cada departamento a cada universidad
df['distancias'] = df.apply(lambda dep: {uni['nombre']: calcular_distancia_cuadras(dep['latitud'], dep['longitud'], uni['latitud'], uni['longitud']) for _, uni in df_universidades.iterrows()}, axis=1)

# Expandir el diccionario de distancias en columnas separadas
distancias_df = df['distancias'].apply(pd.Series)
df = pd.concat([df.drop(columns=['distancias']), distancias_df], axis=1)
```
